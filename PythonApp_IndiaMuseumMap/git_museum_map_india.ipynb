# FINAL - MUSEUMS OF INDIA

Data Source for Museum Names: https://en.wikipedia.org/wiki/List_of_museums_in_India

#### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium as Folium
import geopy as gp
```

#### Loading Data

```python
# Loading data in a Pandas dataframe from Excel File
museum_list= pd.read_excel("Museum_List.xlsx")

# Creating adderss column in dataframe
museum_list["ADDRESS"]= museum_list["MUSEUM NAME"]+", "+museum_list["CITY"]+", "+museum_list["STATE"]
```

#### Final Map Creation and Plotting

```python
# BULK GEOCODING
app= gp.Nominatim(user_agent= 'My_Practice')                           # Instantiating Nominatim
my_data_list= []                                                       # Defining a list to save geocodes of museums

for i in museum_list["ADDRESS"]:                                       # Looping through address column
    if app.geocode(i):
        my_data_list.append(app.geocode(i))                            # Updating geocodes in my_data_list
        
# FINAL MAPPING AND PLOTTING
museum_map= Folium.Map(location= [20,75], zoom_start= 4.49, tiles= "Stamen Terrain")       # Instantiating Folium map
fgm= Folium.FeatureGroup(name= "Museums of India")

for i in my_data_list:                                            # Looping through our geocode list and adding to map
    fgm.add_child(Folium.Marker(location= i[-1], popup= ",".join(i[0].split(",")[1:]), 
                                tooltip= (i[0].split(",")[0])+", "+(i[0].split(",")[-3])))

museum_map.add_child(fgm)
```

```python
museum_map.save("india_museum_map.html")
```
